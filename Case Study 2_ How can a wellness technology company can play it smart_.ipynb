##Adding a new column TotalActiveMinutes to find total active minutes of each umique id member

```python
df_1['TotalActiveMinutes']=df_1['VeryActiveMinutes']+ df_1['FairlyActiveMinutes'] + df_1['LightlyActiveMinutes']
```

**Dropping few irrelevant columns**


```python
df_2=df_1.drop(['TotalDistance','TrackerDistance','VeryActiveDistance','ModeratelyActiveDistance','LightActiveDistance','SedentaryActiveDistance','VeryActiveMinutes','FairlyActiveMinutes','LightlyActiveMinutes','ActivityDate'], axis = 1)
```

```python
df_2
```

**Adding a new column TotalLoggedMinutes to calculate Minutes captured by device during active hours of id member**

```python
df_2['TotalLoggedMinutes']=df_2['SedentaryMinutes'] + df_2['TotalActiveMinutes']
df_2
```

**Total Minutes in a day**

```python
24*60
```

**Finding out how many minutes in a day device was not logging in data**

```python
df_2['TotalUnloggedMinutes']=1440-df_2['TotalLoggedMinutes']
```

```python
df_2
```

**Total Unlogged minutes in a day**

```python
c=df_2[df_2['TotalUnloggedMinutes']!=0]
```

```python
c
```

**Counting the number of times some hours of data was missing**

```python
c['TotalUnloggedMinutes'].count()
```

**Counting uniques ids for which data wasn not logged**

```python
c['Id'].nunique()
```

It means for every member there is some amount of data missing for 31 days.

```python
Id_wise=c.groupby('Id')
```

**Checking for how many days each member had unlogged hours(Minutes) in tracker**

```python
Id_wise.size()
```

```python
new_data=Id_wise.sum()
new_data
```

**Finding out how much data out of 31days(44640min) is missing for each id**

```python
new_data['TotalDaysUnrecorded']=new_data['TotalUnloggedMinutes']/(60*24)
```

```python
new_data['TotalDaysUnrecorded']=new_data['TotalDaysUnrecorded'].round(decimals=2)
```

```python
new_data
```

```python
new_data['TotalDaysUnrecorded'].mean()
```

```python
new_data['UnloggedHours%']=new_data['TotalUnloggedMinutes']*100/(31*24*60)
```

```python
new_data['UnloggedHours%']=new_data['UnloggedHours%'].round(decimals=2)
```

```python
new_data
```

**Average missing data for 31 days span while data was connected in percentage**

```python
new_data['UnloggedHours%'].mean()
```

**Almost 14 hours of data was not logged into the device for every 100 hours in 31 days**
